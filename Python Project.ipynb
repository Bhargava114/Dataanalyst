# Task 2

## Calculating running averages of cost column in Project Data frame

We are using for loop along with the if condition to calculate the running average
At first,we are defining the running_total & Count as 0 at the start of the code.
For every value in for loop, the 'isna' condition checks whether the values are null values/Nan.
If it returns the values as False, we calculate the running total as current running total which is 0 and add the i value. In addition to that, we keep a count of the values
If it returns the value as true, we calculate the running average using the previously calculated running total & count

```python
running_total=0
count=0
for i in range(len(Project_Data_Loaded)):
    if pd.isna(Project_Data_Loaded['Cost'].iloc[i]):
        running_avg = running_total/count
        Project_Data_Loaded['Cost'].iloc[i] = running_avg
    else:
        running_total += Project_Data_Loaded['Cost'].iloc[i]
        count = count+1
Project_Data_Loaded
```
